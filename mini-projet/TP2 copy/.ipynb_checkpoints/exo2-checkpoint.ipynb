```python
def rz_monopolaire_codage(liste_binaire):
    signal_code = []
    
    for i in liste_binaire:
        if i == '1':
            signal_code.extend([5, 0])
        else:
            signal_code.extend([0, 0])
    return signal_code

def rz_monopolaire_decodage(signal_code):
    liste_binaire = ""
    
    for i in range(0, len(signal_code), 2):
        if signal_code[i] == 5:
            liste_binaire += '1'
        else:
            liste_binaire += '0'
    return liste_binaire

# Test de codage et décodage RZ monopolaire
signal_code1 = rz_monopolaire_codage(liste_binaire1)
decoded_signal1 = rz_monopolaire_decodage(signal_code1)

print("Liste binaire originale:", liste_binaire1)
print("Signal codé RZ monopolaire:", signal_code1)
print("Signal décodé RZ monopolaire:", decoded_signal1)
```

```python
def rz_bipolaire_codage(liste_binaire):
    signal_code = []
    
    for i in liste_binaire:
        if i == '1':
            signal_code.extend([5, 0])
        else:
            signal_code.extend([-5, 0])
    return signal_code

def rz_bipolaire_decodage(signal_code):
    liste_binaire = ""
    
    for i in range(0, len(signal_code), 2):
        if signal_code[i] == 5:
            liste_binaire += '1'
        else:
            liste_binaire += '0'
    return liste_binaire

signal_code2 = rz_bipolaire_codage(liste_binaire1)
decoded_signal2 = rz_bipolaire_decodage(signal_code2)

print("Liste binaire originale:", liste_binaire1)
print("Signal codé RZ bipolaire:", signal_code2)
print("Signal décodé RZ bipolaire:", decoded_signal2)
```

```python
def nrz_codage(liste_binaire):
    signal_code = []
    
    for i in liste_binaire:
        if i == '1':
            signal_code.extend([5, 5])
        else:
            signal_code.extend([-5, -5])
    return signal_code

def nrz_decodage(signal_code):
    liste_binaire = ""
    
    for i in range(0, len(signal_code), 2):
        if signal_code[i] == 5:
            liste_binaire += '1'
        else:
            liste_binaire += '0'
    return liste_binaire

signal_code3 = nrz_codage(liste_binaire1)
decoded_signal3 = nrz_decodage(signal_code3)

print("Liste binaire originale:", liste_binaire1)
print("Signal codé NRZ:", signal_code3)
print("Signal décodé NRZ:", decoded_signal3)
```

```python
def nrzi_codage(data):
    encoded_data = []
    prev_bit = 0
    
    if data[0] == '1':
      encoded_data.extend([5, 5])
      prev_bit = 5
    else:
      encoded_data.extend([-5, -5])
      prev_bit = -5
    
    for bit in data[1:]:
        if bit == '0':
            prev_bit = -5 if prev_bit == 5 else 5
        encoded_data.extend([prev_bit, prev_bit])
    
    return encoded_data

def nrzi_decodage(encoded_data):
    decoded_data = []
    prev_bit = '1' if encoded_data[0] == 5 else '0'
    
    for i in range(2, len(encoded_data), 2):
        if encoded_data[i] != encoded_data[i - 1]:
            decoded_data.append('0')
        else:
            decoded_data.append('1')
    
    return ''.join([prev_bit] + decoded_data)

# Test de codage et décodage RZ monopolaire
signal_code4 = nrzi_codage(liste_binaire2)
decoded_signal4 = nrzi_decodage(signal_code4)

print("Liste binaire originale:", liste_binaire2)
print("Signal codé NRZI:", signal_code4)
print("Signal décodé NRZI:", decoded_signal4)
```

```python
def manchester_codage(liste_binaire):
    signal_code = []
    
    for i in liste_binaire:
        if i == '1':
            signal_code.extend([5, -5])
        else:
            signal_code.extend([-5, 5])
    return signal_code

def manchester_decodage(signal_code):
    liste_binaire = ""
    
    for i in range(0, len(signal_code), 2):
        if signal_code[i] == 5:
            liste_binaire += '1'
        else:
            liste_binaire += '0'
    return liste_binaire

signal_code5 = manchester_codage(liste_binaire2)
decoded_signal5 = manchester_decodage(signal_code5)

print("Liste binaire originale:", liste_binaire2)
print("Signal codé manchester:", signal_code5)
print("Signal décodé manchester:", decoded_signal5)
```

```python
def manchester_differentiel_codage(data):
    encoded_data = []
    prev_bit = 0
    
    if data[0] == '1':
      encoded_data.extend([5, -5])
      prev_bit = 5
    else:
      encoded_data.extend([-5, 5])
      prev_bit = -5
    
    for bit in data[1:]:
        if bit == '1':
            prev_bit = 5 if prev_bit == -5 else -5
        encoded_data.extend([prev_bit, -prev_bit])
    
    return encoded_data

def manchester_differentiel_decodage(encoded_data):
    decoded_data = []
    prev_bit = '1' if encoded_data[0] == 5 and encoded_data[1] == -5 else '0'
    
    for i in range(2, len(encoded_data), 2):
        if encoded_data[i] != encoded_data[i - 2]:
            decoded_data.append('1')
        else:
            decoded_data.append('0')
    
    return ''.join([prev_bit] + decoded_data)

# Test de codage et décodage RZ monopolaire
signal_code6 = manchester_differentiel_codage(liste_binaire3)
decoded_signal6 = manchester_differentiel_decodage(signal_code6)

print("Liste binaire originale:", liste_binaire3)
print("Signal codé Manchester différentiel:", signal_code6)
print("Signal décodé Manchester différentiel:", decoded_signal6)
```

```python
def miller_codage(data):
    encoded_data = []
    
    if data[0] == '1':
      encoded_data.extend([5, 5])
    else:
      encoded_data.extend([-5, -5])
    
    for i in range(1, len(data)):
        if data[i] == '0' and data[i - 1] != '0':
          encoded_data.extend([encoded_data[-1], encoded_data[-1]])
        elif data[i] == '0' and data[i - 1] == '0':
          encoded_data.extend([-encoded_data[-1], -encoded_data[-1]])
        else:
          encoded_data.extend([encoded_data[-1], -encoded_data[-1]])
    
    return encoded_data

def miller_decodage(encoded_data):
    decoded_data = ""
    decoded_data += '1' if encoded_data[0] == 5 and encoded_data[1] == -5 else '0'
    
    for i in range(2, len(encoded_data), 2):
        if encoded_data[i] != encoded_data[i - 2] and encoded_data[i] == encoded_data[i + 1]:
            decoded_data += '0'
        elif encoded_data[i] != encoded_data[i + 1]:
            decoded_data += '1'
        else:
            decoded_data += '0'
    
    return decoded_data

# Test de codage et décodage RZ monopolaire
signal_code7 = miller_codage(liste_binaire4)
decoded_signal7 = miller_decodage(signal_code7)

print("Liste binaire originale:", liste_binaire4)
print("Signal codé Miller:", signal_code7)
print("Signal décodé Miller:", decoded_signal7)
```

```python
def cmi_codage(data):
    encoded_data = []
    
    if data[0] == '1':
      encoded_data.extend([-5, 5])
    else:
      encoded_data.extend([-5, -5])
    
    for i in range(1, len(data)):
        if data[i] == '0' and data[i - 1] != '0':
          encoded_data.extend([encoded_data[-1], encoded_data[-1]])
        elif data[i] == '0' and data[i - 1] == '0':
          encoded_data.extend([-encoded_data[-1], -encoded_data[-1]])
        else:
          encoded_data.extend([-encoded_data[-1], encoded_data[-1]])
    
    return encoded_data

def cmi_decodage(encoded_data):
    decoded_data = ""
    for i in range(0, len(encoded_data), 2):
        if encoded_data[i] != encoded_data[i - 2] and encoded_data[i] == encoded_data[i + 1]:
            decoded_data += '0'
        elif encoded_data[i] != encoded_data[i + 1]:
            decoded_data += '1'
        else:
            decoded_data += '0'
    
    return decoded_data

# Test de codage et décodage RZ monopolaire
signal_code8 = cmi_codage(liste_binaire5)
decoded_signal8 = cmi_decodage(signal_code8)

print("Liste binaire originale:", liste_binaire5)
print("Signal codé CMI:", signal_code8)
print("Signal décodé CMI:", decoded_signal8)
```

```python
def index_derniere_valeur_avant_dernier_zero(liste):
    dernier_zero_index = None
    avant_dernier_zero = None

    for i in range(len(liste)):
        if liste[i] == 0:
            if dernier_zero_index is None:
                dernier_zero_index = i
            else:
                avant_dernier_zero = liste[i - 2]
                break

    return liste.index(avant_dernier_zero)
```

```python
def ami_codage(data):
    encoded_data = []
    
    if data[0] == '1':
      encoded_data.extend([5, 5])
    else:
      encoded_data.extend([0, 0])
    
    for i in range(1, len(data)):
        if data[i] == '0':
          encoded_data.extend([0, 0])
        elif encoded_data[index_derniere_valeur_avant_dernier_zero(encoded_data)] == -5:
          encoded_data.extend([5, 5])
        else:
          encoded_data.extend([-encoded_data[-1], encoded_data[-1]])
    
    return encoded_data

def ami_decodage(encoded_data):
    decoded_data = ""
    for i in range(0, len(encoded_data), 2):
        if encoded_data[i] != encoded_data[i - 2] and encoded_data[i] == encoded_data[i + 1]:
            decoded_data += '0'
        elif encoded_data[i] != encoded_data[i + 1]:
            decoded_data += '1'
        else:
            decoded_data += '0'
    
    return decoded_data

# Test de codage et décodage RZ monopolaire
signal_code9 = ami_codage(liste_binaire6)
decoded_signal9 = ami_decodage(signal_code9)

print("Liste binaire originale:", liste_binaire5)
print("Signal codé AMI:", signal_code9)
print("Signal décodé AMI:", decoded_signal9)
```

```python
def mlt3_codage(data):
    encoded_data = []
    
    if data[0] == '1':
      encoded_data.extend([-5, 5])
    else:
      encoded_data.extend([-5, -5])
    
    for i in range(1, len(data)):
        if data[i] == '0' and data[i - 1] != '0':
          encoded_data.extend([encoded_data[-1], encoded_data[-1]])
        elif data[i] == '0' and data[i - 1] == '0':
          encoded_data.extend([-encoded_data[-1], -encoded_data[-1]])
        else:
          encoded_data.extend([-encoded_data[-1], encoded_data[-1]])
    
    return encoded_data

def mlt3_decodage(encoded_data):
    decoded_data = ""
    for i in range(0, len(encoded_data), 2):
        if encoded_data[i] != encoded_data[i - 2] and encoded_data[i] == encoded_data[i + 1]:
            decoded_data += '0'
        elif encoded_data[i] != encoded_data[i + 1]:
            decoded_data += '1'
        else:
            decoded_data += '0'
    
    return decoded_data

# Test de codage et décodage RZ monopolaire
signal_code10 = mlt3_codage(liste_binaire3)
decoded_signal10 = mlt3_decodage(signal_code10)

print("Liste binaire originale:", liste_binaire3)
print("Signal codé MLT3:", signal_code10)
print("Signal décodé MLT3:", decoded_signal10)
```

```python
def deuxBunQ_codage(liste_binaire):
    signal_code = []
    
    for i in range(0, len(liste_binaire), 2):
        if liste_binaire[i:i+2] == '00':
            signal_code.extend([-3, -3])
        elif liste_binaire[i:i+2] == '01':
            signal_code.extend([-1, -1])
        elif liste_binaire[i:i+2] == '11':
            signal_code.extend([1, 1])
        else:
            signal_code.extend([3, 3])
    return signal_code

def deuxBunQ_decodage(signal_code):
    liste_binaire = ""
    
    for i in range(0, len(signal_code), 2):
        if signal_code[i] == -3:
            liste_binaire += '00'
        elif signal_code[i] == -1:
            liste_binaire += '01'
        elif signal_code[i] == 1:
            liste_binaire += '11'
        else:
            liste_binaire += '10'
    return liste_binaire

# Test de codage et décodage RZ monopolaire
signal_code11 = deuxBunQ_codage(liste_binaire7)
decoded_signal11 = deuxBunQ_decodage(signal_code11)

print("Liste binaire originale:", liste_binaire7)
print("Signal codé 2B1Q:", signal_code11)
print("Signal décodé 2B1Q:", decoded_signal11)
```

```python
import matplotlib.pyplot as plt

def plot_signal(y, title):
    time_axis = []  # Créer un axe temporel vide
    new_y = []  # Créer une liste vide pour les valeurs de tension correspondantes
    increment = 0.5  # Définir l'incrément pour avancer les valeurs de x

    # Générer l'axe temporel avec un pas de 0.5 pour chaque paire de valeurs d'ordonnée
    for i, value in enumerate(y):
        time_axis.extend([i * increment, (i + 1) * increment])  # Utiliser l'incrément pour avancer les valeurs de x
        new_y.extend([value, value])  # Répéter la valeur de tension correspondante pour chaque paire d'abscisses

    plt.figure(figsize=(20, 10))  # Définir la taille de la figure à 20 pouces de largeur sur 10 pouces de hauteur
    plt.step(time_axis, new_y, where='mid')  # Modifier cette ligne pour changer where='post' à where='mid' ou where='pre'
    plt.title(title)
    plt.xlabel('Temps bit (Tb)')
    plt.ylabel('Tension (V)')
    plt.ylim(-7, 7)
    
    # Modifier la ligne suivante pour afficher les graduations uniquement par deux valeurs
    plt.xticks([i * increment for i in range(0, len(y) * 2 + 1, 2)])  

    plt.grid(True)
    plt.tight_layout()  # Pour ajuster automatiquement l'espace entre les éléments du graphique
    plt.show()

# Exemple d'utilisation de la fonction
plot_signal(signal_code8, "Exemple de signal avec valeurs de tension")
```
