```python
import mylib as mylib
```

```python
def image_to_binary(path, file_path):
    mod = Image.open(path).mode
    # Vérifier si l'image est en RGB ou en niveaux de gris
    if mod == 'RGB':
        print("L'image chargée est en mode RGB")
        return mylib.RGB_grayscale_binary(path, file_path)
    elif mod == 'L':
        print("L'image chargée est en mode L")
        return mylib.RGB_grayscale_binary(path, file_path)
    elif mod == "1":
        print("L'image chargée est en mode 1")
        return mylib.binary_binary(path, file_path)
    else:
        print("Mode de fichier chargé non reconnu")
        return None

def binary_to_image(binary_sequence, mode, dimensions, traversal_type, output_path):
  img_data = mylib.binary_to(binary_sequence, mode, dimensions, traversal_type)
  
  try:
      # Création de l'image à partir de la matrice
      img = Image.fromarray(img_data, mode=mode)
      img.save(output_path)
      print("Image sauvegardée avec succès.")
  except ValueError as e:
      print("Erreur lors de la conversion de la matrice en image:", e)  
```

```python
def text_to_binary(path):
    with open(path, 'r') as fichier:
        text = fichier.read()
    encoded_text = text.encode()
    
    binary_sequence = ''.join(format(byte, '08b') for byte in encoded_text)
    
    return binary_sequence

def binary_to_text(binary_sequence, output_file_path):
    bytes_array = bytearray()
    for i in range(0, len(binary_sequence), 8):
        byte = binary_sequence[i:i+8]
        bytes_array.append(int(byte, 2))
    
    with open(output_file_path, 'wb') as fichier_text:
        fichier_text.write(bytes_array)
```

```python
def audio_to_binary(path):
    with open(path, 'rb') as fichier_audio:
        donnees_audio = fichier_audio.read()
        liste_binaire = list(donnees_audio)
        binary_sequence = ''.join(format(byte, '08b') for byte in liste_binaire)
    return binary_sequence # Ignorer les erreurs de décodage pour les données non textuelles

def binary_to_audio(binary_sequence, output_file_path):
    bytes_array = bytearray()
    for i in range(0, len(binary_sequence), 8):
        byte = binary_sequence[i:i+8]
        bytes_array.append(int(byte, 2))
    
    with open(output_file_path, 'wb') as fichier_audio:
        fichier_audio.write(bytes_array)
```

```python
binary_sequence = image_to_binary('eren_rgb.jpg', 'l')
print(binary_sequence[0:4000])
binary_to_image(binary_sequence, 'RGB', (224, 222), 'l', 'eren_rgb1.jpg')
```

```python
binary_sequence = image_to_binary('eren_gray.jpg', 'l')
print(binary_sequence[0:4000])
binary_to_image(binary_sequence, 'L', (224, 222), 'l', 'eren_gray1.jpg')
```
