```python
def image_to_binary(path, file_path):
    mod = Image.open(path).mode
    # Vérifier si l'image est en RGB ou en niveaux de gris
    if mod == 'RGB':
        print("L'image chargée est en mode RGB")
        return mb.RGB_grayscale_binary(path, file_path)
    elif mod == 'L':
        print("L'image chargée est en mode L")
        return mb.RGB_grayscale_binary(path, file_path)
    elif mod == "1":
        print("L'image chargée est en mode 1")
        return mb.binary_binary(path, file_path)
    else:
        print("Mode de fichier chargé non reconnu")
        return None

def binary_to_image(binary_sequence, mode, dimensions, traversal_type, output_path):
  img_data = mb.binary_to(binary_sequence, mode, dimensions, traversal_type)
  
  try:
      # Création de l'image à partir de la matrice
      img = Image.fromarray(img_data, mode=mode)
      img.save(output_path)
      print("Image sauvegardée avec succès.")
  except ValueError as e:
      print("Erreur lors de la conversion de la matrice en image:", e)  
```

```python
def rz_monopolaire_codage(liste_binaire):
    signal_code = []
    for bit in liste_binaire:
        if bit == '1':
            signal_code.extend([5, 0])  # Impulsion positive suivie d'une période sans signal
        else:
            signal_code.extend([0, 0])  # Période sans signal pour le bit 0
    return signal_code

def rz_monopolaire_decodage(signal_code):
    liste_binaire = []
    for i in range(0, len(signal_code), 2):
        if signal_code[i] == 5:
            liste_binaire.append(1)  # Si l'échantillon est 1, le bit est 1
        else:
            liste_binaire.append(0)  # Si l'échantillon est 0, le bit est 0
    return liste_binaire

# Test de codage et décodage RZ monopolaire
liste_binaire = "11010001011"
signal_code = rz_monopolaire_codage(liste_binaire)
decoded_signal = rz_monopolaire_decodage(signal_code)

print("Liste binaire originale:", liste_binaire)
print("Signal codé RZ monopolaire:", signal_code)
print("Signal décodé RZ monopolaire:", decoded_signal)
```

```python
import matplotlib.pyplot as plt

def plot_rz_monopolaire(liste_binaire):
    # Création des listes pour les abscisses et les ordonnées
    x = []
    y = []
    
    # Parcours de la liste binaire
    for i, bit in enumerate(liste_binaire):
        # Ajout des valeurs pour chaque échantillon
        x.extend([i, i + 0.5])  # Crée une impulsion pour chaque bit
        if bit == 0:
            y.extend([0, 0])
        else:
            y.extend([5, 0])  # Si le bit est 1, l'impulsion est alternativement positive et négative
    
    # Ajouter le dernier point pour continuer le tracé sur l'axe des abscisses
    x.append(len(liste_binaire))
    y.append(y[-1])  # Garder la dernière valeur de y
    
    # Tracer du graphique
    plt.step(x, y, where='post')
    
    # Ajout des titres et des étiquettes
    plt.title('Codage RZ monopolaire')
    plt.xlabel('Temps bit (Tb)')
    plt.ylabel('Tension (V)')
    
    # Affichage du graphique
    plt.ylim(-7, 7)  # Limite les valeurs de l'axe y entre -1.5 et 1.5
    plt.grid(True)
    plt.show()

# Test de la fonction avec une liste binaire arbitraire
liste_binaire = [1,1,0,1,0,0,0,1,0,1,1]
plot_rz_monopolaire(liste_binaire)
```

```python
def manchester_codage(liste_binaire):
    signal_code = []
    for bit in liste_binaire:
        if bit == 1:
            signal_code.extend([1, -1])  # Pour le bit 0, la transition est de haut en bas
        else:
            signal_code.extend([-1, 1])  # Pour le bit 1, la transition est de bas en haut
    return signal_code

def manchester_decodage(signal_code):
    liste_binaire = []
    for i in range(0, len(signal_code), 2):
        if signal_code[i] == 1 and signal_code[i+1] == -1:
            liste_binaire.append(1)  # Si la transition est de haut en bas, le bit est 0
        elif signal_code[i] == -1 and signal_code[i+1] == 1:
            liste_binaire.append(0)  # Si la transition est de bas en haut, le bit est 1
        else:
            raise ValueError("Signal incorrect : Le codage Manchester a été corrompu.")
    return liste_binaire

# Test de codage et décodage Manchester
liste_binaire = [0,0,1,1,1,0,1,0,0,0,1]
signal_code = manchester_codage(liste_binaire)
decoded_signal = manchester_decodage(signal_code)

print("Liste binaire originale:", liste_binaire)
print("Signal codé Manchester:", signal_code)
print("Signal décodé Manchester:", decoded_signal)
```

```python
import matplotlib.pyplot as plt

# Fonction de codage RZ monopolaire corrigée
def rz_monopolaire_codage(liste_binaire):
    signal = []
    for bit in liste_binaire:
        signal.extend([5, 0] if bit == 1 else [0, 0])
    return signal

# Fonction de décodage RZ monopolaire corrigée
def rz_monopolaire_decodage(signal):
    liste_binaire = []
    for i in range(0, len(signal), 2):
        liste_binaire.append(1 if signal[i] == 5 else 0)
    return liste_binaire

# Fonction de codage RZ bipolaire corrigée selon les nouvelles spécifications
def rz_monopolaire_codage(liste_binaire):
    signal = []
    for bit in liste_binaire:
        signal.extend([5, 0] if bit == 1 else [-5, 0])
    return signal

# Fonction de décodage RZ bipolaire adaptée
def rz_bipolaire_decodage(signal):
    liste_binaire = []
    for i in range(0, len(signal), 2):
        if signal[i] == 5:
            liste_binaire.append(1)
        elif signal[i] == -5:
            liste_binaire.append(0)
    return liste_binaire

# Fonction de codage Manchester inchangée
def manchester_codage(liste_binaire):
    signal = []
    for bit in liste_binaire:
        if bit == 0:
            signal.extend([-5, 5])
        else:
            signal.extend([5, -5])
    return signal

# Fonction de décodage Manchester inchangée
def manchester_decodage(signal):
    liste_binaire = []
    for i in range(0, len(signal), 2):
        if signal[i] == 5 and signal[i+1] == -5:
            liste_binaire.append(1)
        elif signal[i] == -5 and signal[i+1] == 5:
            liste_binaire.append(0)
        else:
            raise ValueError("Signal Manchester invalide")
    return liste_binaire

# Fonction de codage NRZ inchangée
def nrz_codage(liste_binaire):
    signal = []
    for bit in liste_binaire:
        signal.extend([5 if bit == 1 else 0, 5 if bit == 1 else 0])
    return signal

# Fonction de décodage NRZ inchangée
def nrz_decodage(signal):
    liste_binaire = []
    for i in range(0, len(signal), 2):
        liste_binaire.append(1 if signal[i] == 5 else 0)
    return liste_binaire

def plot_signal(y, title):
    time_axis = []  # Créer un axe temporel vide
    new_y = []  # Créer une liste vide pour les valeurs de tension correspondantes

    # Calculer le nombre total de points nécessaires pour représenter tous les bits
    total_pairs = len(y)

    # Générer l'axe temporel pour correspondre à chaque paire de valeurs d'ordonnée
    for i in range(total_pairs):
        time_axis.extend([i + 1, i + 1])  # Répéter l'indice pour chaque paire d'abscisses
        new_y.extend([y[i], y[i]])  # Répéter la valeur de tension correspondante pour chaque paire d'abscisses

    plt.figure(figsize=(20, 10))  # Définir la taille de la figure à 10 pouces de largeur sur 6 pouces de hauteur
    plt.step(time_axis, new_y, where='mid')  # Modifier cette ligne pour changer where='post' à where='mid' ou where='pre'
    plt.title(title)
    plt.xlabel('Temps bit (Tb)')
    plt.ylabel('Tension (V)')
    plt.ylim(-7, 7)
    plt.xticks(range(1, total_pairs + 1))  # Utiliser le nombre total de paires d'abscisses
    plt.grid(True)
    plt.show()



# Exemple d'utilisation
liste_binaire = [1, 1, 0, 1, 0, 0, 0, 1, 0, 1, 1]

signal_rz_monopolaire = rz_monopolaire_codage(liste_binaire)
plot_signal(signal_rz_monopolaire, 'Codage RZ Monopolaire')

signal_rz_bipolaire = rz_bipolaire_codage(liste_binaire)
plot_signal(signal_rz_bipolaire, 'Codage RZ Bipolaire')

signal_manchester = manchester_codage([0, 0, 1, 1, 1, 0, 1, 0, 0, 0, 1])
plot_signal(signal_manchester, 'Codage Manchester')

signal_nrz = nrz_codage(liste_binaire)
plot_signal(signal_nrz, 'Codage NRZ')
```

```python
import matplotlib.pyplot as plt

def rz_monopolaire_encode(data):
    encoded_signal = []
    for bit in data:
        if bit == 1:
            encoded_signal.extend([5] + [0])  # +5V pendant la première moitié, puis 0V pour la seconde moitié
        else:
            encoded_signal.extend([0]*2)  # 0V pendant toute la durée du bit à 0
    return encoded_signal

def rz_monopolaire_decode(signal):
    decoded_data = []
    threshold = 2.5  # Seuil pour décider si le signal est 1 ou 0
    for i in range(0, len(signal), 10):
        if sum(signal[i:i+5]) > threshold:
            decoded_data.append(1)
        else:
            decoded_data.append(0)
    return decoded_data

def plot_encoding(data, encoding_func, title):
    encoded_signal = encoding_func(data)
    time_axis = range(len(encoded_signal))
    plt.step(time_axis, encoded_signal)
    plt.title(title)
    plt.xlabel('Temps (par unité de bit)')
    plt.ylabel('Tension (V)')
    plt.grid(True)
    plt.show()

# Exemple d'utilisation
data = [1, 1, 0, 1, 0, 0, 0, 1, 0, 1, 1]  # Données à encoder
plot_encoding(data, rz_monopolaire_encode, 'RZ Monopolaire Encoding')
```

```python
def vecteur_ligne(matrice):
  vecteur = np.array(matrice, dtype=np.uint8)
  vecteur = vecteur.flatten()
  return vecteur.tolist()

def inverse_vecteur_ligne(vecteur, lignes, colonnes):
    expected_size = lignes * colonnes
    actual_size = len(vecteur)
    
    if expected_size != actual_size:
        raise ValueError(f"Cannot reshape array of size {actual_size} into shape ({lignes},{colonnes})")
    
    mat = np.array(vecteur, dtype=np.uint8).reshape((lignes, colonnes))
    return mat

def vecteur_colonne(matrice):
    vecteur = np.ravel(matrice, order = 'F')
    vecteur = vecteur.tolist()
    return vecteur

def inverse_vecteur_colonne(vecteur, lignes, colonnes):
    expected_size = lignes * colonnes
    actual_size = len(vecteur)
    
    if expected_size!= actual_size:
        raise ValueError(f"Cannot reshape array of size {actual_size} into shape ({lignes},{colonnes})")
    
    mat = np.array(vecteur, dtype=np.uint8).reshape((lignes, colonnes))
    return mat

def vecteur_zigzag(matrice):
    matrice = np.array(matrice, dtype=np.uint8)
    lignes, colonnes = matrice.shape
    solution = [[] for i in range(lignes + colonnes - 1)]
    
    for i in range(lignes):
        for j in range(colonnes):
            somme = i + j
            if somme % 2 == 0:
                solution[somme].insert(0, matrice[i, j])
            else:
                solution[somme].append(matrice[i, j])
    vecteur = np.concatenate(solution)
    
    return vecteur.tolist()

def inverse_vecteur_zigzag(vector, rows, cols):
    mat = [[None] * cols for _ in range(rows)]
    row, col = 0, 0
    direction = 1

    for i in range(rows * cols):
        mat[row][col] = vector[i]
        if direction == 1:
            if col == cols - 1:
                row += 1
                direction = -1
            elif row == 0:
                col += 1
                direction = -1
            else:
                row -= 1
                col += 1
        else:
            if row == rows - 1:
                col += 1
                direction = 1
            elif col == 0:
                row += 1
                direction = 1
            else:
                row += 1
                col -= 1

    return np.array(mat, dtype=np.uint8)

def matrices_vecteur_ligne(matrices):
  # Assuming matrices is of shape (height, width, 3)
    R = matrices[:, :, 0]  # Extract Red channel
    G = matrices[:, :, 1]  # Extract Green channel
    B = matrices[:, :, 2]  # Extract Blue channel
    
    rouge = vecteur_ligne(R)
    vert = vecteur_ligne(G)
    bleu = vecteur_ligne(B)
    
    # Combine the vectors from each channel
    rgb = np.concatenate((rouge, vert, bleu))
    
    return rgb.tolist()

def inverse_vecteur_ligne_matrices(decompressed, L, C):
    if len(decompressed) % 3 != 0:
        raise ValueError("Decompressed data length is not a multiple of 3")

    # Division de la donnée décompressée en trois canaux de couleur
    R = decompressed[:len(decompressed)//3]
    G = decompressed[len(decompressed)//3:2*len(decompressed)//3]
    B = decompressed[2*len(decompressed)//3:]
    
    # Inversion de chaque canal de couleur
    R_inverse = inverse_vecteur_ligne(R, L, C)
    G_inverse = inverse_vecteur_ligne(G, L, C)
    B_inverse = inverse_vecteur_ligne(B, L, C)

    # Création de la matrice tridimensionnelle
    #matrice_tridim = np.array([R_inverse, G_inverse, B_inverse], dtype=np.uint8)
    matrice_tridim = np.stack((R_inverse, G_inverse, B_inverse), axis=-1)
    return np.array(matrice_tridim, dtype=np.uint8)

def matrices_vecteur_colonne(matrices):
  # Assuming matrices is of shape (height, width, 3)
    R = matrices[:, :, 0]  # Extract Red channel
    G = matrices[:, :, 1]  # Extract Green channel
    B = matrices[:, :, 2]  # Extract Blue channel
    
    rouge = vecteur_colonne(R)
    vert = vecteur_colonne(G)
    bleu = vecteur_colonne(B)
    
    # Combine the vectors from each channel
    rgb = np.concatenate((rouge, vert, bleu))
    
    return rgb.tolist()

def inverse_vecteur_colonne_matrices(decompressed, L, C):
    if len(decompressed) % 3 != 0:
        raise ValueError("Decompressed data length is not a multiple of 3")

    # Division de la donnée décompressée en trois canaux de couleur
    R = decompressed[:len(decompressed)//3]
    G = decompressed[len(decompressed)//3:2*len(decompressed)//3]
    B = decompressed[2*len(decompressed)//3:]

    # Inversion de chaque canal de couleur
    R_inverse = inverse_vecteur_colonne(R, L, C)
    G_inverse = inverse_vecteur_colonne(G, L, C)
    B_inverse = inverse_vecteur_colonne(B, L, C)

    # Création de la matrice tridimensionnelle
    #matrice_tridim = np.array([R_inverse, G_inverse, B_inverse], dtype=np.uint8)
    matrice_tridim = np.stack((R_inverse, G_inverse, B_inverse), axis=-1)
    return np.array(matrice_tridim, dtype=np.uint8)

def matrices_vecteur_zigzag(matrices):
    # Assuming matrices is of shape (height, width, 3)
    R = matrices[:, :, 0]  # Extract Red channel
    G = matrices[:, :, 1]  # Extract Green channel
    B = matrices[:, :, 2]  # Extract Blue channel
    
    rouge = vecteur_zigzag(R)
    vert = vecteur_zigzag(G)
    bleu = vecteur_zigzag(B)
    
    # Combine the vectors from each channel
    rgb = np.concatenate((rouge, vert, bleu))
    
    return rgb.tolist()

def inverse_vecteur_zigzag_matrices(decompressed, L, C):
    if len(decompressed) % 3 != 0:
        raise ValueError("Decompressed data length is not a multiple of 3")

    # Division de la donnée décompressée en trois canaux de couleur
    R = decompressed[:len(decompressed)//3]
    G = decompressed[len(decompressed)//3:2*len(decompressed)//3]
    B = decompressed[2*len(decompressed)//3:]

    # Inversion de chaque canal de couleur
    R_inverse = inverse_vecteur_zigzag(R, L, C)
    G_inverse = inverse_vecteur_zigzag(G, L, C)
    B_inverse = inverse_vecteur_zigzag(B, L, C)

    # Création de la matrice tridimensionnelle
    #matrice_tridim = np.array([R_inverse, G_inverse, B_inverse], dtype=np.uint8)
    matrice_tridim = np.stack((R_inverse, G_inverse, B_inverse), axis=-1)
    return np.array(matrice_tridim, dtype=np.uint8)

def matrices_vecteur_ligne_colonne_zigzag(matrices):
  # Assuming matrices is of shape (height, width, 3)
    R = matrices[:, :, 0]  # Extract Red channel
    G = matrices[:, :, 1]  # Extract Green channel
    B = matrices[:, :, 2]  # Extract Blue channel
    
    rouge = vecteur_ligne(R)
    vert = vecteur_colonne(G)
    bleu = vecteur_zigzag(B)
    
    # Combine the vectors from each channel
    rgb = np.concatenate((rouge, vert, bleu))
    
    return rgb.tolist()

def inverse_vecteur_ligne_colonne_zigzag_matrices(decompressed, L, C):
    if len(decompressed) % 3 != 0:
        raise ValueError("Decompressed data length is not a multiple of 3")

    # Division de la donnée décompressée en trois canaux de couleur
    R = decompressed[:len(decompressed)//3]
    G = decompressed[len(decompressed)//3:2*len(decompressed)//3]
    B = decompressed[2*len(decompressed)//3:]

    # Inversion de chaque canal de couleur
    R_inverse = inverse_vecteur_ligne(R, L, C)
    G_inverse = inverse_vecteur_colonne(G, L, C)
    B_inverse = inverse_vecteur_zigzag(B, L, C)

    # Création de la matrice tridimensionnelle
    #matrice_tridim = np.array([R_inverse, G_inverse, B_inverse], dtype=np.uint8)
    matrice_tridim = np.stack((R_inverse, G_inverse, B_inverse), axis=-1)
    return np.array(matrice_tridim, dtype=np.uint8)

def matrices_vecteur_zigzag_colonne_ligne(matrices):
  # Assuming matrices is of shape (height, width, 3)
    R = matrices[:, :, 0]  # Extract Red channel
    G = matrices[:, :, 1]  # Extract Green channel
    B = matrices[:, :, 2]  # Extract Blue channel
    
    rouge = vecteur_zigzag(R)
    vert = vecteur_colonne(G)
    bleu = vecteur_ligne(B)
    
    # Combine the vectors from each channel
    rgb = np.concatenate((rouge, vert, bleu))
    
    return rgb.tolist()

def inverse_vecteur_zigzag_colonne_ligne_matrices(decompressed, L, C):
    if len(decompressed) % 3 != 0:
        raise ValueError("Decompressed data length is not a multiple of 3")

    # Division de la donnée décompressée en trois canaux de couleur
    R = decompressed[:len(decompressed)//3]
    G = decompressed[len(decompressed)//3:2*len(decompressed)//3]
    B = decompressed[2*len(decompressed)//3:]

    # Inversion de chaque canal de couleur
    R_inverse = inverse_vecteur_zigzag(R, L, C)
    G_inverse = inverse_vecteur_colonne(G, L, C)
    B_inverse = inverse_vecteur_ligne(B, L, C)

    # Création de la matrice tridimensionnelle
    #matrice_tridim = np.array([R_inverse, G_inverse, B_inverse], dtype=np.uint8)
    matrice_tridim = np.stack((R_inverse, G_inverse, B_inverse), axis=-1)
    return np.array(matrice_tridim, dtype=np.uint8)
```

```python
from PIL import Image

def RGB_grayscale_vector(image_path, file_path):
    # Ouvrir l'image avec Pillow
    img = Image.open(image_path)

    # Obtenir le mode de l'image
    mode = img.mode
    img = np.array(img, dtype=np.uint8)
    
    if file_path == 'l':
      if mode == 'RGB':
        return matrices_vecteur_ligne(img), img.shape, mode
      elif mode == 'L':
        return vecteur_ligne(img), img.shape, mode
      else:
        print("Mode de fichier non reconnu")
        return None, None, None
    elif file_path == 'c':
      if mode == 'RGB':
        return matrices_vecteur_colonne(img), img.shape, mode
      elif mode == 'L':
        return vecteur_colonne(img), img.shape, mode
      else:
        print("Mode de fichier non reconnu")
        return None, None, None
    elif file_path == 'z':
      if mode == 'RGB':
        return matrices_vecteur_zigzag(img), img.shape, mode
      elif mode == 'L':
        return vecteur_zigzag(img), img.shape, mode
      else:
        print("Mode de fichier non reconnu")
        return None, None, None
    elif file_path == 'lcz':
      if mode == 'RGB':
        return matrices_vecteur_ligne_colonne_zigzag(img), img.shape, mode
      else:
        print("L'image ne peut pas être parcourue en lcz: elle n'est pas en mode RGB")
        return None, None, None
    elif file_path == 'zcl':
      if mode == 'RGB':
        return matrices_vecteur_zigzag_colonne_ligne(img), img.shape, mode
      else:
        print("L'image ne peut pas être parcourue en zcl: elle n'est pas en mode RGB")
        return None, None, None
    else:
      print("Type de parcours inconnue")
      return None, None, None

# Chemin vers l'image que vous souhaitez tester
#image_path = "../TP1/Images/img3/eren_rgb.jpg"
image_path = 'image_binaire.png'
```

```python
def RGB_grayscale_binary(path, file_path):
  img, shape, mode = RGB_grayscale_vector(path, file_path)
  if img == None:
    return None, None, None
  img = ''.join(format(byte, '08b') for byte in img)
  return img, shape, mode
```

```python
def RGB_grayscale_binary_inverse(binary_sequence, path, file_path):
  img, shape, mode = RGB_grayscale_vector(path, file_path)
  if img == None:
    return None, None, None
  bytes_list = [binary_sequence[i:i+8] for i in range(0, len(binary_sequence), 8)]
  vector = bytes(int(byte, 2) for byte in bytes_list)
  return vector, shape, mode
```
