```python
def comparer_et_compter_points(liste_tuples):
    pt1, pt2, points1, points2, total_points, pts = (0,0,0,0,0,0)
    aleatoire = generer_nombres_uniques()
    
    for index in range(len(aleatoire)):
        tup = aleatoire[index]
        ordre_traditionnel = liste_tuples[tup - 1][0]
        versets = liste_tuples[tup - 1][4]
        
        saisie1 = int(input("Quel est l'ordre traditionnel de la sourate " + str(liste_tuples[tup - 1][1]) + " ?: "))
        saisie2 = int(input("Quel est le nombre de versets de la sourate " + str(liste_tuples[tup - 1][1]) + " ?: "))
        
        if saisie1 == ordre_traditionnel:
            points1 += 1
            pt1 = 1
        else:
            print(f"L'ordre traditionnel de la sourate {liste_tuples[tup - 1][2]} n'est pas {saisie1}")
            
        if saisie2 == versets:
            points2 += 1
            pt2 = 1
        else:
            print(f"Le nombre de versets de la sourate {liste_tuples[tup - 1][2]} n'est pas {saisie2}")
        
        pts = pt1 + pt2
        
        print("Total des points pour la sourate " + str(liste_tuples[tup - 1][2]) + " : " + str(pts) + " points au total, " + str(pt1) + " points pour l'ordre traditionnel et " + str(pt2) + " points pour le nombre de versets.")
        total_points = points1 + points2
            
        pt1 = 0
        pt2 = 0
            
        total_points = points1 + points2
    
    print("\nNombre total de points: " + str(total_points) + "/" + str(len(aleatoire) * 2) + " soit " + str((total_points/(len(aleatoire) * 2)) * 100) + "% avec " + str(points1) + "/" + str(len(aleatoire)) + " points en tout pour les ordres traditionnels soit " + str((points1/len(aleatoire)) * 100) + "% et " + str(points2) + "/" + str(len(aleatoire)) + " points pour les nombres de versets soit " + str((points2/len(aleatoire)) * 100) + "%")
    
comparer_et_compter_points(sourates_coran())
```

```python
def caracteres_et_frequences_par_chaine(liste_chaines):
    caracteres_freq_liste = []
    for chaine in liste_chaines:
        caracteres_freq_chaine = []
        freq_dict = {}
        for caractere in chaine:
            if caractere in freq_dict:
                freq_dict[caractere] += 1
            else:
                freq_dict[caractere] = 1
        for caractere, freq in freq_dict.items():
            caracteres_freq_chaine.append((caractere, freq))
        caracteres_freq_liste.append(caracteres_freq_chaine)
    return caracteres_freq_liste

# Exemple d'utilisation
liste_arabe = [tup[1] for tup in sourates_coran()]
resultat = caracteres_et_frequences_par_chaine(liste_arabe)
for item in resultat:
    print(item)
```
